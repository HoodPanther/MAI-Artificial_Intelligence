$$
\begin{align}

the next CVX code corresponds to the optimization of the former problem:

```python
%reset -f
from cvxpy import *
import numpy as np

# Problem data.
m = 30
n = 20
np.random.seed(1)
A = np.random.randn(m, n)
b = np.random.randn(m)

# Construct the problem.
x = Variable(n)

objective = Minimize(sum_squares(A*x - b))
constraints = [norm(x,"inf") <= 1]
prob = Problem(objective, constraints)

# The optimal objective is returned by prob.solve().
result = prob.solve()
# The optimal value for x is stored in x.value.
print(x.value)
# Check the status of the solution (Infeasible, Optimal)
print(prob.status)
```

### A. Understanding the primal
Code the primal problem for hard margin linear SVM (i.e. without slack variables) using CVX. It is recommended to clear your workspace before running the CVX code.

The primal problem hard margin linear SVM is 

$$
\begin{align}
\underset{a,b}{\text{minimize}} & \quad \|a\|_2\\
\text{subject to} & \quad y_i (a^Tx_i+ b) \geq 1, \forall i
\end{align}
$$

```python
# Problem data.
m = len(X)
n = len(X[0])
a = Variable(n)
b = Variable(1)

# Construct the problem.
objective = Minimize(norm(a, 2))
constraints = [mul_elemwise(y,X*a + b) >= 1]
prob = Problem(objective, constraints)

# The optimal objective is returned by prob.solve().
result = prob.solve()
# The optimal value for x is stored in x.value.
print 'a:\n' + str(a.value)
print 'b:\n' + str(b.value)
# Check the status of the solution (Infeasible, Optimal)
print(prob.status)
```

### Answers to Question Block 1
The above plot shows the data set along with the separating hyperplane (solid blue line) and the margins (dashed red lines). The support vectors are the three data points marked in light blue. The support vectors in the hard margin problem are given by those examples i, for which

$$y_i(\mathbf{a}^T\mathbf{x_i}+b)=1$$

Theoretically, that is in an analytical solution, there will always be at least one data point from each class for which the above condition is satisfied (otherwise we can make the margins wider). As we are using an optimization tool here which does not find an exact solution, but a very good approximation thereof, we do not actually find examples for which the value is exactly 1. We therefore decide to round (to 2 decimal places) and consider all those examples support vectors which evaluate very close to 1.

### B. Soft-margin linear SVM

Code the primal problem for the soft-margin linear SVM using CVX. The soft-margin linear SVM is as follows

$$
\begin{align}

\text{subject to} & \quad y_i (a^Tx_i+ b) \geq 1 - \xi_i, \forall i\\
& \quad \xi_i \geq 0
\end{align}
$$

### Answers to Question Block 2:
1) and 2): No, it is not reasonable to assume that the resulting hyperplane should be similar. This is because when using the soft margin formulation with $C=0$, we are still trying to maximise the width of the margin as before, but this time with the loosened constraint $y_i (a^Tx_i+ b) \geq 1 - \xi_i$. As $\xi_i$ can take any positive value and there is no penalisation for large, positive values of $\xi_i$ due to setting $C=0$ in the objective function, the constraint is completely pointless and all we are doing is maximising the width of the margin without any consideration for the data. We conclude that using the soft margin formulation with C=0 is a bad idea as it results in ignoring the data by removing a penalty for wrongly classified examples.
If we wish to obtain a similar result to the hard-margin problem for separable data using the soft margin formulation, then this can be achieved by choosing C very large. This has the effect of strengthening the constraint by highly penalising non-zero values of $\xi_i$ while still maximising the margin. In the limit as C tends to infinity, the solution should converge to the hard margin solution for separable data.

```python
%reset -f
import scipy.io as sio
from cvxpy import *
import numpy as np

data = sio.loadmat('example_dataset_1.mat')
X = data['data'].T
y = data['labels'].astype(float)
```

```python
# Problem data.
m = len(X)
n = len(X[0])

a = Variable(n)
b = Variable(1)
error = Variable(m)
C = [0, 0.01, 1, 100]

a_C = []
b_C = []
for ci in C:
    print 'C: ' + str(ci)
    # Construct the problem.
    objective = Minimize(norm(a, 2) + ci*sum_entries(error))
    constraints = [mul_elemwise(y,X*a + b) >= 1 - error, error >= 0]
    prob = Problem(objective, constraints)

    # The optimal objective is returned by prob.solve().
    result = prob.solve()
    # The optimal value for x is stored in x.value.
    print 'a:\n' + str(a.value)
    print 'b:\n' + str(b.value)
    a_C.append(a.value)    
    b_C.append(b.value)
    # Check the status of the solution (Infeasible, Optimal)
    print(prob.status)
    if ci ==1:
        print 'xi_i (errors):'
        for elemen in error.value:
            print str(round(elemen,3))
```

```python
import matplotlib.pyplot as plt
import matplotlib
colors = ['#FF5733','#9B59B6']

# CLASSIFY
x_draw = np.linspace(-1,1,len(y))

for a,b,c in zip(a_C, b_C, C):    
    plt.figure()    
    plt.scatter(X[:,0], X[:,1], c=y, cmap=matplotlib.colors.ListedColormap(colors))
    plt.hold(True)
    plt.plot(x_draw, (-1/float(a[1]))*(x_draw*float(a[0]) + float(b)), 'b', label=c)
    plt.plot(x_draw, (-1/float(a[1]))*(x_draw*float(a[0]) + float(b) - 1), 'r--')
    plt.plot(x_draw, (-1/float(a[1]))*(x_draw*float(a[0]) + float(b) + 1), 'r--')   
    plt.legend(loc=2)
    # SVM
    SV = []
    for index, (xi, yi, e) in enumerate(zip(X, y, error.value)):
        if round(yi*(xi*a + b), 3) == 1 or round(e,3) > 0:
            SV.append(index)
    for sv in SV:
        plt.scatter(X[sv][0], X[sv][1], color='#48C9B0')
        
plt.show()
```

### Answer to Question Block 4 continued:
4) and 5): The result is not very satisfying: even though the training error is relatively low with $\approx4.9\%$ for a non-separable dataset, 4 out of 13 examples of the critical class have been wrongly classified. This corresponds to $\approx30\%$ of critical examples not being classified as such. We can easily imagine scenarios in which such a high false negative rate is not acceptable, e.g. when -1 examples correspond to a rare desease. We would therefore prefer a classifier which recognises more of the critical data as such, at the cost of misclassifying a few more examples from the non-critical class.   

In general, the cost of an error in a critical class is set by the user/client according to the application. However if we don’t have a priori knowledge it is sensible to use a balancing weight so that an error on the majority class has less importance. A possible balancing weight is to use the quotient between the cardinality of the minority class set over the cardinality of the majority class set.

**HINT: ** We can divide the original SVM optimization problem in the following:

$$
\begin{align}
\underset{a,b}{\text{minimize}} & \quad \|a\|_2 + C (\sum_i u_i + \sum_i v_i),\\
\text{subject to} & \quad a^Tx_i+ b \geq 1 - u_i, \; i: y_i = +1,\\
& \quad a^Tx_i+ b \leq -1 + v_i, \; i: y_i = -1,\\ 
& \quad u_i\geq 0,\\
& \quad v_i\geq 0.
\end{align}
$$

### Answers to Question Block 5:
1) and 2): In order to account for an weighting of the errors, we split up the constraints as:

\begin{align}
& \quad a^Tx_i+ b \geq 1 - u_i, \; i: y_i = +1,\\
& \quad a^Tx_i+ b \leq -1 + v_i, \; i: y_i = -1,\\ 
& \quad u_i\geq 0, \quad v_i\geq 0.
\end{align}

so that $v_i$ represent the errors on the critical class -1. We then include the parameter $W$ in the target function to be minimised to assign more importance to these critical errors:

\begin{align}
\underset{a,b}{\text{minimize}} & \quad \|a\|_2 + C \big(\sum_i u_i + W \sum_i v_i\big)\\
\text{where} & \quad W= \frac{|Cl_{+1}|}{|Cl_{-1}|}
\end{align}

and $|Cl_{+1}|,|Cl_{-1}|$ are the sizes of the two classes. In our case $W=\frac{110}{13}\approx 8.5$, so that one critical error has the same penalty as roughly 8.5 non-critical errors.

```python
%reset -f
import scipy.io as sio
from cvxpy import *
import numpy as np

data = sio.loadmat('example_dataset_3.mat')
X = data['data'].T
y = data['labels'].astype(float)
```

```python
import pandas as pd

D2 = pd.DataFrame(X)
D2['y'] = y

len_positive = D2.loc[D2['y'] == 1, 'y'].size
len_negative = D2.loc[D2['y'] == -1, 'y'].size
W = float(len_positive)/float(len_negative)
index_positive = D2.loc[D2['y'] == 1, 'y'].index
index_negative = D2.loc[D2['y'] == -1, 'y'].index
```

```python
# Problem data.
m = len(X)
n = len(X[0])

a = Variable(n)
b = Variable(1)
u = Variable(len_positive)
v = Variable(len_negative)
C = [0.01, 0.1, 1, 10, 100, 1000, 10000]

a_C = []
b_C = []
misclassifiedsmallclass=[]
misclassifiedbigclass=[]
for ci in C:
    # Construct the problem.
    objective = Minimize(norm(a, 2) + ci* (sum_entries(u) + W*sum_entries(v)) )
    constraints = [X[index_positive]*a + b >= 1 - u, X[index_negative]*a + b <= -1 + v, u >= 0, v >= 0]    
    prob = Problem(objective, constraints)

    # The optimal objective is returned by prob.solve().
    result = prob.solve()
    b_C.append(b.value)
    a_C.append(a.value)   
      
    # TRAINING ERRORS
    successsmall=0
    successbig=0
    for xi, yi in zip(X, y):
        decision = 1 if xi * a.value + b.value > 0 else -1
        if decision == yi == -1:
            successsmall += 1
        if decision == yi == +1:
            successbig +=1
    misclassifiedsmallclass.append(len_negative-successsmall)
    misclassifiedbigclass.append(len_positive-successbig)

print 'C values: ' + str(C)
print 'Critical errors:' + str(misclassifiedsmallclass)
print 'Non-critical errors:' + str(misclassifiedbigclass)
```
